```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from yellowbrick.regressor import ResidualsPlot 
import matplotlib.pyplot as plt
```

```python
#Carregando os dados
dados = pd.read_csv('/content/dados.csv')
dados
```

```python
#Separando os valores e atribuindo a variáveis
x = dados.iloc[:, 3].values
X = pd.DataFrame(x, columns = ['Tempo_adesão'])
y = dados.iloc[:, 4].values
Y = pd.DataFrame(y, columns = ['Valor_total'])
```

```python
#Chamando a função Regressão Linear
x = x.reshape(-1, 1)
modelo = LinearRegression()
modelo.fit(x, y)
```

```python
#Juntando os dados
dados1 = pd.concat([X, Y], axis = 1)
dados1
```

```python
#Vizualizando as previsões
previsao = modelo.predict(x)
previsoes = pd.DataFrame(previsao, columns = ['Previsto'])
previsoes
```

```python
#Juntando os dados em uma tabela
comparaçao = pd.concat([X, Y, previsoes], axis = 1)
comparaçao
```

```python
#Calculando o erro (diferença entre a previsão do modelo e o valor real)
Erro = (Y.values - previsoes.values)
Erro = pd.DataFrame(Erro, columns = ['Erro'])
comparaçao = pd.concat([comparaçao, Erro], axis = 1)
comparaçao
```

```python
#Calculando o erro absoluto
erro_abs = abs(Y.values - previsoes.values)
erro_absoluto = pd.DataFrame(erro_abs, columns = ['Erro_absoluto'])
comparaçao = pd.concat([comparaçao, erro_absoluto], axis = 1)
comparaçao
```

#MEA (Mean Error Absolute) = 37,39

```python
a = np.mean(comparaçao.iloc[:, 4].values)
print('O erro médio absoluto é: {:.2f}'.format(a))
```

#RMSE (Root Mean Squared Error) = 46,57

```python
qd = comparaçao['Erro_absoluto']**2
qd
rmse = (qd.sum()/500)**(1/2)
print('A raiz quadrada do erro médio é: {:.2f}'.format(rmse))
```

```python
#Exibindo o erro absoluto em percentual
Erro_percentual_abs = (comparaçao['Erro_absoluto'] / comparaçao['Valor_total'])*100
Erro_percentual_abs = pd.DataFrame(Erro_percentual_abs, columns = ['Erro (%)'])
comparaçao = pd.concat([comparaçao, Erro_percentual_abs], axis = 1)
comparaçao
```

#MAPE (Mean Absolute Percentage Error) = 7,78%

```python
a = np.mean(comparaçao['Erro_percentual'])
print(' O erro médio percentual absoluto é: {:.2f}'.format(a))
```
