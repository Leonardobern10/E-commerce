#Analisando as vendas

```python
venda = dados['Total_gasto_anual']
print('Os clientes geraram uma receita de R${:.2f}'.format(venda.sum()))
print('A empresa recebeu em média R${:.2f} por venda'.format(venda.mean()))
print('O valor da maior receita foi R${:.2f}'.format(venda.max()))
print('O valor da menor venda foi R${:.2f}'.format(venda.min()))
```
