#Entendendo padrões de compras

Acesso ao aplicativo

```python
aplicativo = dados['Tempo_app']
print('Em média os clientes permanecem por {:.3f} minutos no app'.format(aplicativo.mean()))
print('O acesso mais rádido durou {:.3f} minutos'.format(aplicativo.min()))
print('O acesso mais demorado durou {:.3f} minutos'.format(aplicativo.max()))
media = np.mean(aplicativo)
```

```python
site = dados['Tempo_site']
print('Em média os clientes permanecem por {:.3f} minutos no site'.format(site.mean()))
print('O acesso mais rápido durou {:.3f} minutos'.format(site.min()))
print('O acesso mais memorado durou {:.3f} minutos'.format(site.max()))
```

```python
tp_adesao = dados['Tempo_de_adesao']
print('Em média os clientes demoram {:.3f} minutos para concluir a compra'.format(tp_adesao.mean()))
print('A compra mais rápida durou {:.3f} minutos'.format(tp_adesao.min()))
print('A compra mais memorada durou {:.3f} minutos'.format(tp_adesao.max()))
```
