```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import norm
```

```python
dados = pd.read_csv('/content/dados.csv')
dados
```

```python
del dados ['Unnamed: 0']
```

```python
aplicativo = dados['Tempo_app']
site = dados['Tempo_site']
tp_adesao = dados['Tempo_de_adesao']
```

#Verificando a distribuição dos valores

Acesso ao aplicativo

```python
sns.displot(dados, x = aplicativo, height = 4, aspect = 3)
```

```python
fig, ax = plt.subplots()
stats.probplot(aplicativo, fit = True, plot = ax)
plt.show()
```

```python
statistic, p_valor = stats.shapiro(aplicativo)
print('O p-Value = {:.2f}'.format(p_valor))
```

```python
sns.boxplot(x = aplicativo, dodge = True)
```

Acesso ao site

```python
sns.displot(dados, x = site, height = 4, aspect = 3)
```

```python
fig, ax = plt.subplots()
stats.probplot(site, fit = True, plot = ax)
plt.show()
```

```python
statistic, p_valor = stats.shapiro(site)
print('O p-Value = {:.2f}'.format(p_valor))
```

```python
sns.boxplot(x = site, dodge = True)
```
